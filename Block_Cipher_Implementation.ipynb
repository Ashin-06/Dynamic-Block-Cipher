```python
# To generate random key, random subkey and to store it

def generaterandomkey(length):
    key = ""
    for _ in range(length):
        key += random.choice("ABCDEFGHIJKLMNOPQRSTUVWXYZ!@#%&^*+=-_:;<>.,?/0123456789")
    return key

def generaterandomkey1(length):
    key1 = ""
    for _ in range(length):
        key1 += random.choice(key)
    return key1

def generaterandomsubkeys(key, numsubkeys, a):
    subkeys = []
    for _ in range(numsubkeys):
        subkeys.append(generaterandomkey1(a))
    return subkeys

def storesubkeysinstring(subkeys):
    subkeysstring = ""
    for subkey in subkeys:
        subkeysstring += subkey
    return subkeysstring
```

```python
# To convert the subkeys to binary and store it in a list
# Also inverting the list of subkeys for decryption
# Entering the number of rounds for both encryption and decryption

subkey2 = ['']*len(subkeys)
subkey1 = subkeys
for i in range(len(subkeys)):
    for char in subkeys[i]:
        if char in nos:
            subkey2[i] += nos[char]
        else :
            subkey2[i] = char    

revsubkey = subkey2[::-1]


numrounds = 32

```

```python
# Function to do the encryption, decryption, xor

def roundingnetwork(block, numrounds, subkey):
    
    for item in subkey:
        
       
        block2 = str(block)
        partlength = len(block2)//4
        subblock1, subblock2 = block[:partlength], block[partlength:2*partlength]
        subblock3 =  block[2*partlength: 3*partlength] 
        subblock4 =  block[3*partlength:]
       
        
        
        for round in range(numrounds):
            newsubblock3 = xor(subblock4 ,item)
            newsubblock2 = xor(subblock3 ,item)
            newsubblock1 = xor(subblock2 ,item)
            newsubblock4 = xor(subblock1 ,item)
    
            subblock11, subblock22, subblock33, subblock44 = newsubblock1, newsubblock2, newsubblock3, newsubblock4
        
            if (len(subblock11)<1024):
                paddedsubblock1 = f"{subblock11:1024}"
            else:
                paddedsubblock1 = subblock11
            if (len(subblock22)<1024):
                paddedsubblock2 = f"{subblock22:1024}"
            else:
                paddedsubblock2 = subblock22
            if (len(subblock33)<1024):
                paddedsubblock3 = f"{subblock33:1024}"
            else:
                paddedsubblock3 = subblock33
            if (len(subblock44)<1024):
                paddedsubblock4 = f"{subblock44:1024}"
            else:
                paddedsubblock4 = subblock44
                
            subblock1, subblock2, subblock3, subblock4 =  paddedsubblock1, paddedsubblock2, paddedsubblock3, paddedsubblock4
        return subblock1+subblock2+subblock3+subblock4

def xor(a, b):
    return ''.join([str(int(x) ^ int(y)) for x, y in zip(a, b)])

```
